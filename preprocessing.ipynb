```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_colwidth', None)
```

```python
data_train = pd.read_csv('data/train.csv')
data_train.head()
```

```python
# we want classes to start from 0 
print(data_train["Class Index"].unique())

data_train["Class"] = data_train["Class Index"].apply(lambda x: x-1)

print(data_train["Class"].unique())
```

```python
# we dont want two text features
data_train["Text"] = data_train["Title"] + " " + data_train["Description"]
data_train.head()
```

```python
from bs4 import BeautifulSoup
import html
import re

def clean_html(text: str)-> str:
    soup = BeautifulSoup(text, "html.parser")
    text = html.unescape(soup.get_text())
    soup = BeautifulSoup(text, "html.parser")
    text = html.unescape(soup.get_text()).strip('\\').strip()
    return text

def lowercasing(text: str):
    return text.lower()

data_train["Text"] = data_train["Text"].apply(clean_html)
data_train["Text"] = data_train["Text"].apply(lowercasing)
```

```python
from tqdm import tqdm
 
def process_duplicates(df: pd.DataFrame) -> pd.DataFrame:
    """If duplicate instances have the same target class leave one in dataframe, else delete all of them"""
    duplikati = df[df.duplicated(subset=['Text'], keep=False)]

    for indeks, row in tqdm(duplikati.iterrows()):
        if(indeks not in duplikati.index): 
            continue

        isti_tekst_razlicite_labele = df[(df['Text'] == row['Text']) & (df['Class'] != row['Class'])]
        isti_tekst_iste_labele = df[(df['Text'] == row['Text']) & (df['Class'] == row['Class'])]
        isti_tekst_iste_labele.drop(indeks) 
        indices_to_drop = [] 

        if not isti_tekst_razlicite_labele.empty:
            indices_to_drop.extend(isti_tekst_razlicite_labele.index.tolist())
            indices_to_drop.append(indeks)
        
        if not isti_tekst_iste_labele.empty:
            indices_to_drop.extend(isti_tekst_iste_labele.index.tolist())
            indices_to_drop.remove(indeks)

        indices_to_drop = [idx for idx in indices_to_drop if idx in df.index]
        df = df.drop(indices_to_drop)
        duplikati=duplikati.drop(indices_to_drop)

    return df

data_train = process_duplicates(data_train)
```

```python
import string

def remove_punctuation(text: str) -> str:
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', string.punctuation))

def clean_agency(text: str) -> str:
    # pattern that deletes city and news agency from the start of the news
    pattern = r'^[A-Z,/\s]+(?:\([A-Z.]+\)\s*-\s*|:\s*)' 
    return re.sub(pattern, '', text, flags=re.IGNORECASE)

def clean_agency2(text: str) -> str:
    pattern = r'^[^-]*-\s*'
    return re.sub(pattern, '', text)

def clean_title(text: str) -> str:
    pattern = r'\s*\([^\)]*\)$'
    return re.sub(pattern, '', text)

def preprocess_pipeline(df: pd.DataFrame) -> pd.DataFrame:
    df["Description"] = df["Description"].apply(clean_agency)
    df["Description"] = df["Description"].apply(clean_agency2)
    df["Title"] = df["Title"].apply(clean_title)

    df["Class"] = df["Class Index"].apply(lambda x: x-1)
    df["Text"] = df["Title"] + " " + df["Description"]

    df["Text"] = df["Text"].apply(clean_html)
    df["Text"] = df["Text"].apply(lowercasing)
    df["Text"] = df["Text"].apply(remove_punctuation)

    return process_duplicates(df)

train = preprocess_pipeline(pd.read_csv("data/train.csv"))
test = preprocess_pipeline(pd.read_csv("data/test.csv"))
```

```python
train.to_csv("data/preprocessed_train.csv")
test.to_csv("data/preprocessed_test.csv")
```
