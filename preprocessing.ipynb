```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_colwidth', None)
```

```python
data_train = pd.read_csv('data/train.csv')
data_train.head()
```

```python
# we want classes to start from 0 
print(data_train["Class Index"].unique())

data_train["Class"] = data_train["Class Index"].apply(lambda x: x-1)

print(data_train["Class"].unique())
```

```python
# we dont want two text features 
data_train["Text"] = data_train["Title"] + " " + data_train["Description"]
data_train.head()
```

```python
# daj primer pre i posle i zasto se neki korak sproveo
def lowercasing(text: str):
    return text.lower()

def get_tag_index(text:str, tag: str)->int:
    return len(text) if tag not in text else text.index(tag)

def remove_html(text: str) -> str:
    return text[:get_tag_index(text, "&lt")]

data_train["Text"] = data_train["Text"].apply(remove_html)
data_train["Text"] = data_train["Text"].apply(lowercasing)
```

```python
from tqdm import tqdm
# promeni funkciju da mogu da se proslede imena kolona 

def obradi_duplikate(df):
    """Ako duplikati imaju istu labelu, ostavi jedan, ako nemaju obrisi sve"""
    duplikati = df[df.duplicated(subset=['Text'], keep=False)]

    for indeks, row in tqdm(duplikati.iterrows()):
        if(indeks not in duplikati.index): #zato sto ih brisem i tokom petlje
            continue

        isti_tekst_razlicite_labele = df[(df['Text'] == row['Text']) & (df['Class'] != row['Class'])]
        isti_tekst_iste_labele = df[(df['Text'] == row['Text']) & (df['Class'] == row['Class'])]
        isti_tekst_iste_labele.drop(indeks) 
        indices_to_drop = [] 

        if not isti_tekst_razlicite_labele.empty:
            indices_to_drop.extend(isti_tekst_razlicite_labele.index.tolist())
            indices_to_drop.append(indeks)
        
        if not isti_tekst_iste_labele.empty:
            indices_to_drop.extend(isti_tekst_iste_labele.index.tolist())
            indices_to_drop.remove(indeks)

        indices_to_drop = [idx for idx in indices_to_drop if idx in df.index]
        df = df.drop(indices_to_drop)
        duplikati=duplikati.drop(indices_to_drop)

    return df

data_train = obradi_duplikate(data_train)
```

```python
# proveriti vaznost ostalih koraka emotikona, emojija, stemovanja i lematizacije

def preprocess_pipeline(df: pd.DataFrame):
    df["Class"] = df["Class Index"].apply(lambda x: x-1)
    df["Text"] = df["Title"] + " " + df["Description"]

    df["Text"] = df["Text"].apply(remove_html)
    df["Text"] = df["Text"].apply(lowercasing)

    return obradi_duplikate(df)

train = preprocess_pipeline(pd.read_csv("data/train.csv"))
test = preprocess_pipeline(pd.read_csv("data/test.csv"))
```

```python
train.to_csv("data/preprocessed_train.csv")
test.to_csv("data/preprocessed_test.csv")
```
